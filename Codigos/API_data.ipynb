````python
import requests
import pandas as pd
import time

# CryptoCompare API URL
url = 'https://min-api.cryptocompare.com/data/pricehistorical'

# Parameters
params = {
    'api_key': api_key,
    'limit': 2000,  # maximum limit
    'calculationType': 'Close'
}

# Create an empty DataFrame to store the results
df = pd.DataFrame()

# Iterate over the trading pairs list and make a request to the CryptoCompare API for each pair
for trading_pair in trading_pairs:

    # Set the fsym and tsym parameters
    params['fsym'] = trading_pair
    params['tsyms'] = 'USD'

    # Get the current timestamp
    end_timestamp = int(time.time())

    # Convert the start year to a Unix timestamp
    start_timestamp = int(time.mktime(time.strptime("2014-01-01", "%Y-%m-%d")))

    prices = []
    dates = []

    while end_timestamp > start_timestamp:
        # Set the toTs parameter
        params['toTs'] = end_timestamp

        # Make the request
        response = requests.get(url, params=params)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:

            # Get the JSON response
            data = response.json()

            # Add the data to the prices list
            prices.append(data[trading_pair]['USD'])

            # Add the timestamp to the dates list
            dates.append(end_timestamp)

            # Update the end_timestamp for the next request
            end_timestamp -= 86400  # Subtract one day in seconds

    # Create a DataFrame for the current trading pair
    df_pair = pd.DataFrame({
        'Date': pd.to_datetime(dates, unit='s'),
        trading_pair: prices
    })

    # Merge the DataFrame with the main DataFrame
    if df.empty:
        df = df_pair
    else:
        df = pd.merge(df, df_pair, on='Date', how='outer')

# Print the DataFrame
print(df)
````
